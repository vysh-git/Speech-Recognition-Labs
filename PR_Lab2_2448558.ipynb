```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Helper Functions ----------

def plot_dtft(signal, n, title="DTFT Spectrum"):
    """Compute and plot DTFT using freqz (continuous frequency spectrum)."""
    from scipy.signal import freqz
    w, h = freqz(signal, worN=2048, whole=True)
    plt.figure(figsize=(10,4))
    plt.plot(w/np.pi, np.abs(h))  # Normalize freq axis
    plt.title(title)
    plt.xlabel("Normalized Frequency (x Ï€ rad/sample)")
    plt.ylabel("Magnitude")
    plt.grid(True)
    plt.show()

def plot_dft(signal, title="DFT Spectrum"):
    """Compute and plot DFT using FFT (discrete frequency spectrum)."""
    N = len(signal)
    X = np.fft.fft(signal)
    freqs = np.fft.fftfreq(N, d=1/fs)  # frequency axis
    plt.figure(figsize=(10,4))
    plt.stem(freqs[:N//2], np.abs(X[:N//2]))
    plt.title(title)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.grid(True)
    plt.show()

# ---------- Global Parameters ----------
fs = 1000   # Sampling frequency (Hz)
t = np.arange(0, 1, 1/fs)  # Time axis (1 second duration)
n = np.arange(len(t))      # Sample index
```

```python
# (a) Generate sine wave (5 Hz)
f = 5
x = np.sin(2*np.pi*f*t)

# (b) Plot time-domain
plt.figure(figsize=(10,4))
plt.plot(t, x)
plt.title("Sinusoidal Signal (5 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# (c) DTFT
plot_dtft(x, n, title="DTFT of 5 Hz Sinusoidal")

# (d) DFT
plot_dft(x, title="DFT of 5 Hz Sinusoidal")
```

```python
# (a) Generate composite: 5 Hz + 20 Hz + 50 Hz
x_comp = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*20*t) + 0.8*np.sin(2*np.pi*50*t)

# (b) Time-domain
plt.figure(figsize=(10,4))
plt.plot(t, x_comp)
plt.title("Composite Signal (5 Hz + 20 Hz + 50 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# (c) DTFT
plot_dtft(x_comp, n, title="DTFT of Composite Signal")

# (d) DFT
plot_dft(x_comp, title="DFT of Composite Signal")
```

```python
# (a) Generate exponentially decaying signal
x_exp = np.exp(-3*t)  # decay rate = 3

# (b) Time-domain
plt.figure(figsize=(10,4))
plt.plot(t, x_exp)
plt.title("Exponentially Decaying Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# (c) DTFT
plot_dtft(x_exp, n, title="DTFT of Exponential Signal")

# (d) DFT
plot_dft(x_exp, title="DFT of Exponential Signal")

# (e) Analysis: Longer decay (slower in time) â†’ narrower frequency spread
print("Observation: The exponential decay shortens the time-domain energy, causing wider frequency spread.")
```

```python
# (a) Rectangular pulse (between 0.2s and 0.4s)
x_rect = np.zeros_like(t)
x_rect[(t >= 0.2) & (t <= 0.4)] = 1

# (b) Time-domain
plt.figure(figsize=(10,4))
plt.plot(t, x_rect)
plt.title("Rectangular Pulse (0.2s - 0.4s)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# (c) DTFT
plot_dtft(x_rect, n, title="DTFT of Rectangular Pulse")

# (d) DFT
plot_dft(x_rect, title="DFT of Rectangular Pulse")

# (e) Analysis: Rectangular in time â†’ sinc-shaped spectrum in frequency
print("Observation: Rectangular time-domain signal corresponds to a sinc-shaped frequency spectrum.")
```
